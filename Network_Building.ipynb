```python
import datetime
import pandas as pd
import numpy as np
from collections import defaultdict
from mediadata import get_headline_df, get_biden_trump_dataframes

start_date = datetime.date(2024, 3, 1)
end_date = datetime.date(2024, 4, 1)

headlines = defaultdict(list)
media_orgs = ['CNN', 'FOX', 'MSNBC']

for media_org in media_orgs:
    
    current = start_date
    
    while current < end_date:
        try:
            df = get_headline_df(media_org, current)
            headlines[media_org].append(df)
        except:
            pass
        
        current += datetime.timedelta(1)
        
        
    
```

```python
headlines_comb = {}
for key, value in headlines.items():
    headlines_comb[key] = pd.concat(value)
```

```python
headlines_biden = {}
headlines_trump = {}
for key, value in headlines_comb.items():
    biden, trump, _, _ = get_biden_trump_dataframes(value)
    headlines_biden[key] = biden
    headlines_trump[key] = trump
    
```

```python
all_biden_headlines = set()
for value in headlines_biden.values():
    for headline in value['Headline']:
        all_biden_headlines.add(headline)
        
```
